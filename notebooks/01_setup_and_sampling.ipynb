```python
import pandas as pd
RAW_FILE = "application_train.csv"
DATA_PATH = r"C:\Users\felix\hci_final_project\data\raw"
SAMPLE_PATH = r"C:\Users\felix\hci_final_project\data\sample"
df_head = pd.read_csv(f"{DATA_PATH}\\{RAW_FILE}", nrows=5, low_memory=False)
df_head.columns.tolist()[:12], df_head.shape
```

```python
TARGET = "TARGET"  
pd.read_csv(f"{DATA_PATH}\\{RAW_FILE}", usecols=[TARGET])[TARGET].value_counts(normalize=True)
    
```

Checking TARGET proportion because of imbalance

```python
full = pd.read_csv(f"{DATA_PATH}\\{RAW_FILE}", low_memory=False)
miss = full.isna().mean().sort_values(ascending=False).head(30)
const_cols = [c for c in full.columns if full[c].nunique(dropna=False)==1]
miss, const_cols[:10]
```

Calculating column with most NaN

```python
pos_total = int(full[TARGET].sum())
min_pos = 1000                             
frac = min(0.10, max(0.01, min_pos / max(pos_total, 1)))  
frac
```

sampling with 0.10 and 0.01

```python
sample = full.groupby(TARGET, group_keys=False).sample(frac=frac, random_state=42)
sample_ids = set(sample["SK_ID_CURR"])
sample.to_csv("C:\\Users\\felix\\hci_final_project\\data\\sample\\application_train_sample.csv", index=False)
```
